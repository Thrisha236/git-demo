An AI agent that reminds students to study based on exam date, subjects, and available study hours, and gives warnings if time is insufficient.

Use Case:



Smart Study Reminder / Planner Agent

 Problem Statement:

Students often forget to study regularly and fail to manage time before exams, leading to poor performance.

Objective:

To build an AI agent that:



Accepts exam date, subjects, and daily study hours

Calculates remaining days

Suggests a study plan

Gives reminders and warnings

```python
from datetime import datetime
```

```python
exam_date = input("Enter exam date (YYYY-MM-DD): ")
subjects = input("Enter subjects (comma separated): ")
daily_hours = float(input("Enter daily study hours: "))
```

```python
exam_date = datetime.strptime(exam_date, "%Y-%m-%d")
today = datetime.today()

remaining_days = (exam_date - today).days
subject_list = subjects.split(",")
```

```python
if remaining_days <= 0:
    print("Exam date has already passed!")
else:
    print(f"Days left for exam: {remaining_days}")

    hours_per_subject = (remaining_days * daily_hours) / len(subject_list)

    print("\n Study Plan:")
    for subject in subject_list:
        print(f"- {subject.strip()}: {round(hours_per_subject, 2)} hours")
```

```python
if daily_hours < 2:
    print("\n Warning: Daily study time is too low. Increase study hours!")
else:
    print("\n You are on track. Keep studying regularly!")
```

```python
print("\nDaily Study Reminder:")
print("Reminder: Study consistently today to stay on track for your exam!")
```
